```python
import json
import requests
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
# load datase
df = pd.read_csv('https://raw.githubusercontent.com/pplonski/datasets-for-start/master/adult/data.csv', skipinitialspace=True)
```

```python
x_cols = [c for c in df.columns if c != 'income']

X = df[x_cols]
y = df['income']

# show first rows of data
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
for i in range(100):
    input_data = dict(X_test.iloc[i])
    target = y_test.iloc[i]
    
    r = requests.post("https://127.0.0.1:8000/api/v1/income_classifier/predict?status=ab_testing", input_data)
    response = r.json()
    # provide feedback  
    requests.post("https://127.0.0.1:8000/api/v1/mlrequests/{}".format(response['request_id'], {"feedback": target}))
```
